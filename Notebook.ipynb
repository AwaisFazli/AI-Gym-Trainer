```python
def calculate_angle(a,b,c):
    a = np.array(a) # First
    b = np.array(b) # Mid
    c = np.array(c) # End
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle >180.0:
        angle = 360-angle
        
    return int(angle) 

def calculate_distance(a, b):
    x1, y1 = a
    x2, y2 = b
    
    x1 = x1*w
    x2 = x2*w
    
    y1 = y1*h
    y2 = y2*h
    
    
    dist = m.sqrt((x2-x1)**2+(y2-y1)**2)
    return dist
```
